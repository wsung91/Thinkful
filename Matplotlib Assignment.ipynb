## Data Visualization of Avocado Prices Data on Kaggle

```python
#import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load csv file as pandas df
with open('avocado.csv') as file:
    avocado_df = pd.read_csv(file)
    
avocado_df.info() 
```

```python
#average price
avo_us_df = avocado_df.loc[avocado_df.region == 'TotalUS']
avo_us_df.info()
avo_us_df.sort_values(by='Date')
```

```python
#Filter Data for Conventional/Organic
avo_us_conv_df = avo_us_df.loc[avo_us_df.type == 'conventional']
avo_us_org_df = avo_us_df.loc[avo_us_df.type == 'organic']
```

```python
#need help with changing size, font of matplotlib plot 
plt.figure()
plt.plot_date(avo_us_conv_df.Date, avo_us_conv_df.AveragePrice, color='blue')
plt.plot_date(avo_us_org_df.Date, avo_us_org_df.AveragePrice, color = 'green')
plt.ylabel('Price in $')
plt.title('Avocado Price Since 2015')
plt.show()
```

- From this chart you can see that the price of avocados reached its peak sometime in 2017, maybe due to lack of supply or a sudden increase in demand, also organic avocados are generally more expensive than conventional avocados

```python
#group conventional by year
avo_us_conv_df['year'].value_counts()
```

```python
#get sum of total volume groupby year
conv_total = avo_us_conv_df.groupby('year', as_index=False).agg({"Total Volume": "sum"})
type(conv_total)
```

```python
#get sum by PLU code
conv_sm_Hass_total =avo_us_conv_df.groupby('year', as_index=False).agg({"4046": "sum"})
conv_lg_Hass_total = avo_us_conv_df.groupby('year', as_index=False).agg({"4225": "sum"})
conv_xl_Hass_total = avo_us_conv_df.groupby('year', as_index=False).agg({"4770": "sum"})
```

```python
#plot year and total volume on a bar chart
plt.bar(conv_total.year, conv_total['Total Volume']/1000000)
plt.bar(conv_sm_Hass_total.year, conv_sm_Hass_total['4046']/1000000, color='green')
plt.bar(conv_lg_Hass_total.year, conv_lg_Hass_total['4225']/1000000, color='orange')
plt.bar(conv_xl_Hass_total.year, conv_xl_Hass_total['4770']/1000000, color='purple')
plt.xlabel('YEAR')
plt.xticks(conv_total.year)
plt.ylabel('Avocados in Million')
plt.show()
```

- Total volume avocados sold have been increasing last couple of years, of the Hass avocados, more large Hass Avocados are sold than any other Hass avocados

```python
#group by region
plt.figure(figsize=(25,10))
conv_region = avocado_df.groupby('region', as_index=False).agg({"AveragePrice": "mean"})
plt.bar(conv_region.region, conv_region.AveragePrice)
plt.xlabel('Region Where Avocados were Sold at', fontsize=30)
plt.xticks(rotation=90, fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('Price in $', fontsize=30)
plt.show()
```

The avocado's price ranged from around $1 to 1.75. By average, avocados were most expensive in Hartford/Springfield area as well as in San Francisco area and it was the chepeast in Houston area
